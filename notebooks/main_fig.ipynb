```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# set width of bar
barWidth = 0.02
fig = plt.subplots(figsize =(6, 7))
 
# set height of bar
Train_acc = [0, 0.994232, 0.994232, 0]
Test_acc = [0, 0.416667, 0.488889, 0]
 
# Set position of bar on X axis
br1 = np.arange(len(Train_acc))
br1 = [0.175, 0.2, 0.3, 0.325]
br2 = [x + barWidth for x in br1]
br3 = [x + barWidth for x in br2]

# Make the plot
plt.bar(br1, Train_acc, color ='b', alpha=0.7, width = barWidth,
        edgecolor ='black', linewidth=1, label ='Train')
plt.bar(br2, Test_acc, color ='r', alpha=0.6, width = barWidth,
        edgecolor ='black', linewidth=1, label ='Test')

# Adding configs
plt.ylabel('Worst-group Accuracy',  fontsize = 25)
plt.xticks([r + barWidth * 0.5 for r in br1[1:3]], 
           ['standard', 'importance sampling', ''],fontsize=6)
plt.title('SGD', fontweight ='bold',  fontsize = 22, y=1.01)
# plt.xticks(rotation=90)
plt.rc('xtick', labelsize=14) 
# plt.legend(fontsize=20, bbox_to_anchor=(1, 0.6))
plt.ylim(0.0, 1.01)
plt.tick_params(axis='x', labelsize=17)
plt.tick_params(axis='y', labelsize=15)
plt.grid(axis='y', linestyle='--')
plt.savefig('main_fig1.pdf', bbox_inches='tight')


```
